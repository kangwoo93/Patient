```python
# Function to calculate age based on birth date
def calculate_age(birth_date_str):
    birth_date = datetime.strptime(birth_date_str, '%Y-%m-%d')
    today = datetime.today()
    age = today.year - birth_date.year - ((today.month, today.day) < (birth_date.month, birth_date.day))
    return age

# Initialize dictionaries to store scores for each res_code
score_dict = defaultdict(lambda: defaultdict(int))
score_dict_repres = defaultdict(lambda : defaultdict(int))
score_dict_hourwages = defaultdict(lambda : defaultdict(int))
score_dict_gender = defaultdict(lambda: defaultdict(int))
score_dict_age = defaultdict(lambda: defaultdict(int))

# Loop through each row in res_df to get the pre_location, pre_gender, and pre_age preferences for each res_code
for index, row in res_df.iterrows():
    res_code = row['res_code']
    pre_locations = [row['pre_location_1'], row['pre_location_2'], row['pre_location_3']]
    pre_repres = [row['pre_repre_1'], row['pre_repre_2'], row['pre_repre_3']]
    pre_hourwages = [row['pre_hourwage_1'], row['pre_hourwage_2'], row['pre_hourwage_3']]
    pre_gender = row['pre_gender']
    pre_ages = [row['pre_age_1'], row['pre_age_2'], row['pre_age_3']]
    
    # Loop through each row in care_df to calculate the scores based on location, gender, and age preferences
    for idx, caregiver_row in care_df.iterrows():
        g_name = caregiver_row['g_name']
        
        g_locations = [caregiver_row['g_location1'], caregiver_row['g_location2'], caregiver_row['g_location3']]
        g_services = [caregiver_row['g_service1'], caregiver_row['g_service2'], caregiver_row['g_service3']]
        g_hourwages = [caregiver_row['g_hourwage1'], caregiver_row['g_hourwage2'], caregiver_row['g_hourwage3']]
        g_gender = caregiver_row['g_gender']
        g_birth = caregiver_row['g_birth']
        
        # Calculate age
        g_age = calculate_age(g_birth)
        
        # Calculate the scores based on location
        score = 0
        if pre_locations[0] == g_locations[0]:
            score += 10
        if pre_locations[0] == g_locations[1]:
            score += 8
        if pre_locations[0] == g_locations[2]:
            score += 6
        if pre_locations[1] == g_locations[0]:
            score += 7
        if pre_locations[1] == g_locations[1]:
            score += 5
        if pre_locations[1] == g_locations[2]:
            score += 3
        if pre_locations[2] == g_locations[0]:
            score += 4
        if pre_locations[2] == g_locations[1]:
            score += 2
        if pre_locations[2] == g_locations[2]:
            score += 1
        score_dict[res_code][g_name] = score

        
        # Calculate the scores based on service
        repres_score = 0
        if pre_repres[0] == g_services[0]:
            repres_score += 10
        if pre_repres[0] == g_services[1]:
            repres_score += 8
        if pre_repres[0] == g_services[2]:
            repres_score += 6
        if pre_repres[1] == g_services[0]:
            repres_score += 7
        if pre_repres[1] == g_services[1]:
            repres_score += 5
        if pre_repres[1] == g_services[2]:
            repres_score += 3
        if pre_repres[2] == g_services[0]:
            repres_score += 4
        if pre_repres[2] == g_services[1]:
            repres_score += 2
        if pre_repres[2] == g_services[2]:
            repres_score += 1
        score_dict_repres[res_code][g_name] = repres_score
        
        
        # Calculate the scores based on hourwage
        hourwages_score = 0
        if pre_hourwages[0] == g_hourwages[0]:
            hourwages_score += 10
        if pre_hourwages[0] == g_hourwages[1]:
            hourwages_score += 8
        if pre_hourwages[0] == g_hourwages[2]:
            hourwages_score += 6
        if pre_hourwages[1] == g_hourwages[0]:
            hourwages_score += 7
        if pre_hourwages[1] == g_hourwages[1]:
            hourwages_score += 5
        if pre_hourwages[1] == g_hourwages[2]:
            hourwages_score += 3
        if pre_hourwages[2] == g_hourwages[0]:
            hourwages_score += 4
        if pre_hourwages[2] == g_hourwages[1]:
            hourwages_score += 2
        if pre_hourwages[2] == g_hourwages[2]:
            hourwages_score += 1
        score_dict_hourwages[res_code][g_name] = hourwages_score
        
        
        
        # Calculate the scores based on gender
        gender_score = 0
        if pre_gender == 'M' and g_gender == 'M':
            gender_score += 5
        if pre_gender == 'W' and g_gender == 'W':
            gender_score += 5
        if pre_gender == 'M' and g_gender == '상관없음':
            gender_score += 3
        if pre_gender == '상관없음' and g_gender == 'M':
            gender_score += 3
        if pre_gender == '상관없음' and g_gender == 'W':
            gender_score += 3
        if pre_gender == 'W' and g_gender == '상관없음':
            gender_score += 3
        if pre_gender == 'M' and g_gender == 'W':
            gender_score += 1
        if pre_gender == 'W' and g_gender == 'M':
            gender_score += 1
        score_dict_gender[res_code][g_name] = gender_score

        # Calculate the scores based on age
        age_score = 0
        if pre_ages[0] in str(g_age):
            age_score += 5
        if pre_ages[1] in str(g_age):
            age_score += 3
        if pre_ages[2] in str(g_age):
            age_score += 1
        score_dict_age[res_code][g_name] = age_score

# Re-calculate the weighted scores based on rank preferences
final_weighted_score_dict = defaultdict(lambda: defaultdict(int))

# Loop through each row in res_df to get the preferences and ranks for each res_code
for index, row in res_df.iterrows():
    res_code = row['res_code']
    rank1 = row['rank1']
    rank2 = row['rank2']
    rank3 = row['rank3']
    rank4 = row['rank4']
    rank5 = row['rank5']
    
    # Determine the weights based on ranks
    weights = {'선호지역': 1, '선호성별': 1, '선호나이대': 1, '선호서비스':1, '시급':1}  # Initialize weights to 1
    if rank1 == '선호지역':
        weights['선호지역'] = 0.35
    if rank1 == '선호성별':
        weights['선호성별'] = 0.35
    if rank1 == '선호나이대':
        weights['선호나이대'] = 0.35
    if rank1 == '선호서비스':
        weights['선호서비스'] = 0.35
    if rank1 == '시급':
        weights['시급'] = 0.35
        
    if rank2 == '선호지역':
        weights['선호지역'] = 0.25
    if rank2 == '선호성별':
        weights['선호성별'] = 0.25
    if rank2 == '선호나이대':
        weights['선호나이대'] = 0.25
    if rank2 == '선호서비스':
        weights['선호서비스'] = 0.25
    if rank2 == '시급':
        weights['시급'] = 0.25
        
    if rank3 == '선호지역':
        weights['선호지역'] = 0.17
    if rank3 == '선호성별':
        weights['선호성별'] = 0.17
    if rank3 == '선호나이대':
        weights['선호나이대'] = 0.17
    if rank3 == '선호서비스':
        weights['선호서비스'] = 0.17
    if rank3 == '시급':
        weights['시급'] = 0.17
        
    if rank4 == '선호지역':
        weights['선호지역'] = 0.13
    if rank4 == '선호성별':
        weights['선호성별'] = 0.13
    if rank4 == '선호나이대':
        weights['선호나이대'] = 0.13
    if rank4 == '선호서비스':
        weights['선호서비스'] = 0.13
    if rank4 == '시급':
        weights['시급'] = 0.13
        
    if rank5 == '선호지역':
        weights['선호지역'] = 0.1
    if rank5 == '선호성별':
        weights['선호성별'] = 0.1
    if rank5 == '선호나이대':
        weights['선호나이대'] = 0.1
    if rank5 == '선호서비스':
        weights['선호서비스'] = 0.1
    if rank5 == '시급':
        weights['시급'] = 0.1
    
    
    # Calculate the final weighted score for each g_name based on the given ranks and weights
    for g_name in score_dict[res_code].keys():  
        location_score = score_dict[res_code][g_name] * weights['선호지역']
        repres_score = score_dict_repres[res_code][g_name] * weights['선호서비스']
        hourwages_score = score_dict_hourwages[res_code][g_name] * weights['시급']
        gender_score = score_dict_gender[res_code][g_name] * weights['선호성별']
        age_score = score_dict_age[res_code][g_name] * weights['선호나이대']
        
        # Calculate the final weighted score
        final_weighted_score = location_score + repres_score + hourwages_score + gender_score + age_score
        
        # Store the final weighted score for this res_code and g_name
        final_weighted_score_dict[res_code][g_name] = final_weighted_score

# Find the top 3 g_names for each res_code based on the final weighted scores
final_top_3_weighted_recommendations = {}
for res_code, scores in final_weighted_score_dict.items():
    sorted_scores = sorted(scores.items(), key=lambda x: x[1], reverse=True)[:3]
    final_top_3_weighted_recommendations[res_code] = [g_name for g_name, score in sorted_scores]

final_top_3_weighted_recommendations
```
